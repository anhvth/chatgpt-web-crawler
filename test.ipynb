```python
import pandas as pd
```

```python
import io


def input_preproc(user):
    from speedy_utils import jloads
    return jloads(user.split('##')[2])[1]


def output_preproc(assistant):
    assistant = assistant.replace(" ", "\n")
    assistant = pd.read_csv(
        io.StringIO(assistant),
    )
    return assistant


def preproc(row):
    user = input_preproc(row["user"])
    assistant = output_preproc(row["assistant"])
    user = pd.DataFrame(user)
    return user, assistant
```

```python
from speedy_utils import *
input = pd.read_csv('/tmp/prompt.csv')
data = '/tmp/prompt_submited_response.csv'
df = pd.read_csv(data)
assistants = []
for i in range(len(df)):
    row = df.iloc[i]
    
    assistants.append(output_preproc(row["assistant"]))

records = []
for _, row in input.iterrows():
    _records = jloads(row['records'])
    records += _records

# records = flatten_list(records)
records = pd.DataFrame(records)
assistants = pd.concat(assistants)


# join assistants and records by RowID
df = pd.merge(records, assistants, on='RowID', how='left')
```
